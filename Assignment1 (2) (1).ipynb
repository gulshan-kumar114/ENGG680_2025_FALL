```python
# step 5
# define a curve fitting function named fit_polynomial_curve
# the function should implement the linear least square estimation
# the function should include all important intermediate steps in LSE
# the input and output of the funciton are described in the comment below in red.
# please keep code clean and well-commented
# (7 marks)

import numpy as np

def fit_polynomial_curve(a: np.ndarray, l: np.ndarray, P: np.ndarray, deg: int) -> tuple:
    '''
    Args:
        a:              function input (constants)
        l:              noisy observations
        P:              observation weight matrix
        deg:            degree of polynomial to fit to data
    Returns:
        x:              estimated polynomial coefficients
        cov_x:          uncertainty in estimated coefficients
        apv:            a posterier variance factor (goodness of fit)
        v:              observation residuals
    '''
    n = len(l)
    A = np.vander(a, deg+1, increasing=True)
    At_p = A.T @ P
    N = At_p @ A
    n_vec = At_p @ l
    x = np.linalg.solve(N, n_vec)
    l_hat = A @ x
    v = l - l_hat
    dof = n - (deg + 1)
    apv = (v.T @ P @ v) / dof
    cov_x = apv * np.linalg.inv(N)

    return (x, cov_x, apv, v)
```

```python
# step 6
# call the fit_polynomial_curve function and fetch the results
# use polynomial degrees of: 2, 4, 6
# (1 marks)
degrees= [2, 4, 6]
coefficients_dict = {}
covariance_dict = {}
apv_dict = {}
residuals_dict = {}
for deg in degrees:
  x, Cov_x, apv, v = fit_polynomial_curve(a, l, P, deg)
  coefficients_dict[deg] = x
  covariance_dict[deg] = Cov_x
  apv_dict[deg] = apv
  residuals_dict[deg]= v

  print(f"Degree {deg} Polynomial:")
  print("coefficients (x):", x)
  print("A Posteriori Variance Factor (apv):", apv)
  print("Residuals (first 5 values):", v[:5])
  print("-"*50)
```

```python
# step 7
# use the estimated coefficients x and constants array a, construct polynomial functions for each degree
# hint: use polyval function from numpy library (check input argument carefully)
# (1 marks)

quadratic_uw = np.polyval(coefficients_dict[2][::-1], a)
print("quadratic fit (First 5 values):",quadratic_uw[:5])
quartic_uw =np.polyval(coefficients_dict[4][::-1], a)
print("quartic fit(first 5 values):", quartic_uw[:5])
hexic_uw = np.polyval(coefficients_dict[6][::-1], a)
print("hexic fit (first 5 values):", hexic_uw[:5])
```

```python
# step 8
# use the matplotlib to plot each polynomial function, noisy observations l, true observations l_true, together in one single figure
# for the polynomials of line, cubic and quintic, use color of orange, green and red respectively.
# (2 marks)
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))

# Scatter plot for noisy observations
ax.scatter(a, l, color='k', label='f(a)')
# Scatter plot for True observations
ax.scatter(a,l_true,color='tab:blue',label='f=(a)+e',marker='o',alpha=0.3)
ax.plot(a, quadratic_uw, color='orange', label='quadratic fit (deg=2)', linewidth=2)
ax.plot(a, quartic_uw, color='green', label='quartic fit (deg=4)', linewidth=2)
ax.plot(a, hexic_uw, color='red', label='hexic fit (deg=6)', linewidth=2)
ax.set_xlabel('a')
ax.set_ylabel('observations')
ax.set_title('True and Noisy Observations and Polynomial Fits')
ax.set_xlim(-1,1)
ax.legend()
ax.grid(True, which='both',ls='-', color="0.65")

fig.tight_layout()
plt.show()
```

```python
# step 9
# print the a posterier variance factor for each polynomial function
# (3 marks)

for deg in [2, 4, 6]:
  print(f"Degree {deg} Polynomial: A Posteriori Variance Factor (apv) = {apv_dict[deg]:.6f}")
```

```python
# step 10
# compute and print the norm between the predicted observations (output from each polynomial function) and the true observations
# (3 marks)
norm_quadratic_uw = np.linalg.norm(quadratic_uw - l_true)
norm_quartic_uw = np.linalg.norm(quartic_uw - l_true)
norm_hexic_uw = np.linalg.norm(hexic_uw - l_true)
print("Norm for quadratic fit:", norm_quadratic_uw)
print("Norm for quartic fit:", norm_quartic_uw)
print("Norm for hexic fit:", norm_hexic_uw)

```

```python
# step 11
# comment on the results of a posterier variance factor and norm, what pattern do you observe

# Pattern observed-
# 1. apv decreases with polynomial degree because higher-degree polynomial fit the noisy data better
# 2. Norm decreases initially but may slightly increase if overfitting occurs
# 3. The best polynomial degree is often the one with low norm but not the lowest apv.
```

```python
# step 12
# now, let's define the P matrix with the proper weights of observation
# use the variance of observation l_var, define P matrix
# (1 mark)
P = np.diag(1 / l_var)
print("weighted observation matrix P:\n",P)
```
